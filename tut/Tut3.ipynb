```python
import pandas as pd
import numpy as np
import statistics as st
import matplotlib.pyplot as plt
import pylab 
import scipy.stats as scst

from statistics import mean
from statistics import median
from statistics import variance
data = pd.read_csv(r"C:/Data/gasoline.csv")
print(data)
```

```python
######################  QQ PLOTS
scst.probplot(data["y"], dist="norm", plot=pylab)
pylab.title('QQ Plot') 
pylab.show()
```

```python
#################################  HISTOGRAMS:
x = data['y']
inter = (max(x) - min(x) )/(len(x)) 
l = list(np.arange(min(x),max(x),step = inter))
y = scst.norm.pdf(l,loc = mean(x),scale = np.std(x)) 

plt.plot(l, y)
plt.hist(data['y'], bins=None, range=None, density=True, color='C2')
plt.title('Histogram of Gasoline Mileage') 
plt.xlabel('') 
plt.ylabel('Probability') 
plt.show()
```

```python
# separate observations by group

auto = data.y[data.x11 == 1 ]
#print(len(auto))
print([len(auto), min(auto), max(auto)])



manual = data.y[data.x11 == 0 ]
#print(len(manual))
print(  [len(manual),min(manual), max(manual)] )
```
