```python
data = pd.read_csv(r"/Users/zzy/Desktop/bike_day.csv", sep = ",")
print(data)
```

```python
#Q1:
cnt_work = 0
cnt_work_list = [];
cnt_no_work = 0
cnt_no_work_list = [];
for i in range(len(data.cnt)):
    if (data.workingday[i] == 1):
        cnt_work_list.append(data.cnt[i])
        cnt_work += data.cnt[i]
        
    elif (data.workingday[i] == 0):
        cnt_no_work += data.cnt[i]
        cnt_no_work_list.append(data.cnt[i])

print(cnt_work/ len(data.cnt))#average rental count on working day.
print(cnt_no_work/ len(data.cnt))#average rental count on non- working day.

        
```

```python

#Q2:


work = cnt_work_list

inter = (max(work) - min(work) )/(len(work)) 
l = list(np.arange((min(work)),(max(work)),step = inter))
y = scst.norm.pdf(l,loc = mean(work),scale = np.std(work))

plt.plot(l, y)
plt.hist(work, bins=None, range=None, density=True, color='C5')
plt.title('Histogram of work') 
plt.xlabel('') 
plt.ylabel('Probability') 
plt.show()
```
